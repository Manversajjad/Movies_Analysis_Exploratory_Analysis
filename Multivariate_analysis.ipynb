```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Heat-Map :- Heat maps utilize color intensity to help us visualize the frequency counts for categorical data.**

```python
tips = sns.load_dataset("tips")
```

Showing the 5 Entries from the tips dataset using Head()

```python
tips.head()
```

Pivot table helps to create a table from long info to wide info


```python
# Assuming `tips` is your DataFrame
tips_pivoted = pd.pivot_table(
    data=tips,
    index='day',
    columns='time',
    values='total_bill',
    aggfunc='count'  # Using 'count' instead of len for better compatibility
)

# Display the pivoted table
print(tips_pivoted)
```

```python
ax = sns.heatmap(data = tips_pivoted)
```

```python
ax = sns.heatmap(data = tips_pivoted, cmap='Greens')
```

Using Annotation

```python
ax = sns.heatmap(data = tips_pivoted, cmap='Greens', annot=True)
```

Scatter Plot :- Used to compare the two variables and plot them.

```python
scat_tips = sns.scatterplot(data = tips, x = "total_bill", y="tip", color = 'green')
```

Plotting the same data using regplot to display the data statistics with the trend line

```python
scat_tips = sns.regplot(data = tips, x = "total_bill", y="tip", color = 'green')
```
